```python
# Function to clean text data
def clean_text(text):
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'\n', ' ', text)  # Remove newlines
    text = re.sub(r'\[.*?\]', '', text)  # Remove text in brackets
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words with numbers
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

# Apply cleaning to text columns
text_columns = ['Job Description', 'skills','Job Title']
for col in text_columns:
    df[col] = df[col].apply(clean_text)

# Check the cleaned data
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Function to calculate Jaccard similarity for skill matching
def jaccard_similarity(skills1, skills2):
    set1, set2 = set(skills1.split()), set(skills2.split())
    return len(set1 & set2) / len(set1 | set2) if set1 | set2 else 0

# Sample a subset of job posts to speed up computations
jobpost_sample = df.sample(n=min(1000, len(df)), random_state=42)  # Adjust sample size

# Normalize salary and experience differences
scaler = MinMaxScaler()

# Compute skill similarity, experience score, and salary compatibility
interaction_matrix = pd.DataFrame(index=jobSeeker.index, columns=jobpost_sample.index)

for seeker_id, seeker in jobSeeker.iterrows():
    for job_id, job in jobpost_sample.iterrows():
        # Skills Match (60% weight)
        skill_match = jaccard_similarity(seeker["skills"], job["skills"])

        # Experience Score (20% weight) - Closer is better
        experience_diff = abs(seeker["experience"] - job["experience"])
        exp_score = 1 / (1 + experience_diff)  # Normalize

        # Salary Compatibility (20% weight)
        min_salary, max_salary = job["min_salary"], job["max_salary"]
        salary_score = 1 if min_salary <= seeker["expected_salary"] <= max_salary else 0

        # Compute weighted sum
        interaction_score = (0.6 * skill_match) + (0.2 * exp_score) + (0.2 * salary_score)
        interaction_matrix.loc[seeker_id, job_id] = interaction_score

# Convert matrix to float
interaction_matrix = interaction_matrix.astype(float)
interaction_matrix.head()
print("hello")
```

```python
from scipy.sparse.linalg import svds

# Assuming interaction_matrix is already computed
# Apply Singular Value Decomposition (SVD)
k = min(50, min(interaction_matrix.shape) - 1)  # Number of latent factors
U, sigma, Vt = svds(interaction_matrix, k=k)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)

# Reconstruct the predicted interaction matrix
predicted_matrix = np.dot(np.dot(U, sigma), Vt)

# Convert back to a DataFrame
predicted_df = pd.DataFrame(predicted_matrix, 
                            index=interaction_matrix.index, 
                            columns=interaction_matrix.columns)

# Show the top predicted job recommendations for each user
predicted_df.head()
```

```python
import pandas as pd
import random
import string

# Given set of skills
skills_set = {'ai', 'sql', 'analysis', 'ml', 'azure', 'flask', 'docker', 'java', 'linux', 'tensorflow', 'expressjs', 
              'c', 'graphql', 'react', 'data', 'aws', 'cloud', 'nodejs', 'typescript', 'fastapi', 'css', 'python',
              'kubernetes', 'html', 'javascript', 'pytorch', 'django'}

# Function to generate a random name
def random_name():
    return ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase, k=8))

# Function to generate a random email
def random_email(name):
    domains = ['gmail.com', 'yahoo.com', 'outlook.com', 'example.com']
    return f"{name.lower()}{random.randint(100,999)}@{random.choice(domains)}"

# Function to generate a random password
def random_password():
    return ''.join(random.choices(string.ascii_letters + string.digits, k=10))

# Function to generate random skills
def random_skills():
    return ' '.join(random.sample(list(skills_set), random.randint(2, 6)))  # Select 2-6 random skills

# Function to generate random experience name
def random_experience():
    positions = ['Software Engineer', 'Data Scientist', 'ML Engineer', 'Backend Developer', 'Frontend Developer', 
                 'Cloud Engineer', 'AI Specialist', 'DevOps Engineer']
    return random.choice(positions)

# Function to generate random location
def random_location():
    locations = ['New York', 'San Francisco', 'Los Angeles', 'Austin', 'Seattle', 'Boston', 'Chicago', 'Denver']
    return random.choice(locations)

# Generate 100,000 job seekers
data = []
for i in range(1, 100001):
    name = random_name()
    email = random_email(name)
    data.append([
        i,  # JobSeeker ID
        name,
        email,
        random_password(),
        'JobSeeker',
        round(random.uniform(30000, 150000), 2),  # Salary Expectation
        random.randint(0, 15),  # Experience in years
        random_skills(),
        random_location(),
        random_experience()
    ])

# Create DataFrame
df = pd.DataFrame(data, columns=['jobseeker_id', 'name', 'email', 'password', 'userType', 'salaryExpectation',
                                 'experience', 'skills', 'location', 'position'])

# Save to CSV
df.to_csv('jobseekers_dataset.csv', index=False)

print("Dataset created successfully: jobseekers_dataset.csv")
```

```python
import numpy as np
import pandas as pd

# Load the saved model
model_data = np.load("../model/svd_model.npz", allow_pickle=True)

# Extract stored matrices
U = model_data["U"]
sigma = model_data["sigma"]
Vt = model_data["Vt"]

# Extract job and user indices
user_index = model_data["user_index"]
job_index = model_data["job_index"]

# Reconstruct the predicted interaction matrix
predicted_matrix = np.dot(np.dot(U, sigma), Vt)

# Convert back to DataFrame
predicted_df = pd.DataFrame(predicted_matrix, index=user_index, columns=job_index)

# Function to get top-N recommended job post IDs for a given user
def get_top_n_jobs(user_id, N=5):
    if user_id not in predicted_df.index:
        return f"User ID {user_id} not found in the dataset."
    
    # Get the user's predicted scores for all jobs
    user_predictions = predicted_df.loc[user_id]
    
    # Get the top-N job IDs with highest scores
    top_job_ids = user_predictions.nlargest(N).index.tolist()
    
    return top_job_ids

# Example usage
user_id = 2  # Replace with an actual job seeker ID
top_jobs = get_top_n_jobs(user_id, N=5)
print("Top recommended job IDs:", top_jobs)
```
