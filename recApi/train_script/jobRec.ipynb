```python
import numpy as np
import pandas as pd
import re
import string
```

```python
df = pd.read_csv("../dataset/job_descriptions.csv")
df.head()
```

```python
df.shape
```

```python
columns_to_drop = [
    "Qualifications", "Country", "longitude", "latitude", "Work Type",
    "Company Size", "Contact", "Role", "Job Portal", "Benefits",
    "Responsibilities", "Company", "Company Profile","Preference","Contact Person","location","Job Posting Date"
]

df = df.drop(columns=columns_to_drop, errors='ignore')  # 'errors="ignore"' prevents errors if a column is missing

print(df.shape)  
df.head()
```

```python
# Extract min and max experience from the 'experience' column
df[['min_experience', 'max_experience']] = df['Experience'].str.extract(r'(\d+)\s*to\s*(\d+)')

# Convert to numeric (optional)
df['min_experience'] = df['min_experience'].astype(float)
df['max_experience'] = df['max_experience'].astype(float)

# Drop the original 'experience' column
df = df.drop(columns=['Experience'])

# Check the result # Verify the changed
df = df.drop(columns=['max_experience'])
df.head()
```

```python
# Extract min and max salary from the 'Salary Range' column
df[['min_salary', 'max_salary']] = df['Salary Range'].str.extract(r'(\d+)K?-?(\d+)K?')

# Convert to numeric (optional)
df['min_salary'] = df['min_salary'].astype(float) * 1000  # Convert 'K' to full value
df['max_salary'] = df['max_salary'].astype(float) * 1000  # Convert 'K' to full value

# Drop the original 'Salary Range' column
df = df.drop(columns=['Salary Range'])

# Check the result
df.head()
```

```python
df = df.rename(columns={'min_experience': 'experience'})
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['min_salary', 'max_salary']] = scaler.fit_transform(df[['min_salary', 'max_salary']])

# Check the result
df.head()
```

```python
# Function to clean text data
def clean_text(text):
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'\n', ' ', text)  # Remove newlines
    text = re.sub(r'\[.*?\]', '', text)  # Remove text in brackets
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words with numbers
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

# Apply cleaning to text columns
text_columns = ['Job Description', 'skills','Job Title']
for col in text_columns:
    df[col] = df[col].apply(clean_text)

# Check the cleaned data
df.head()
```

```python
df[['experience']] = scaler.fit_transform(df[['experience']])
df.head()
```

```python
jobSeeker = pd.read_csv("../dataset/job_seekers.csv")
jobSeeker.head()
```

```python
# Apply cleaning to text columns
jobSeeker['skills'] = jobSeeker['skills'].apply(clean_text)

# Check the cleaned data
jobSeeker.head()
```

```python
jobSeeker[['expected_salary']] = scaler.fit_transform(jobSeeker[['expected_salary']])
jobSeeker.head()
```

```python
jobSeeker[['experience']] = scaler.fit_transform(jobSeeker[['experience']])
jobSeeker.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Function to calculate Jaccard similarity for skill matching
def jaccard_similarity(skills1, skills2):
    set1, set2 = set(skills1.split()), set(skills2.split())
    return len(set1 & set2) / len(set1 | set2) if set1 | set2 else 0

# Sample a subset of job posts to speed up computations
jobpost_sample = df.sample(n=min(1000, len(df)), random_state=42)  # Adjust sample size

# Normalize salary and experience differences
scaler = MinMaxScaler()

# Compute skill similarity, experience score, and salary compatibility
interaction_matrix = pd.DataFrame(index=jobSeeker.index, columns=jobpost_sample.index)

for seeker_id, seeker in jobSeeker.iterrows():
    for job_id, job in jobpost_sample.iterrows():
        # Skills Match (60% weight)
        skill_match = jaccard_similarity(seeker["skills"], job["skills"])

        # Experience Score (20% weight) - Closer is better
        experience_diff = abs(seeker["experience"] - job["experience"])
        exp_score = 1 / (1 + experience_diff)  # Normalize

        # Salary Compatibility (20% weight)
        min_salary, max_salary = job["min_salary"], job["max_salary"]
        salary_score = 1 if min_salary <= seeker["expected_salary"] <= max_salary else 0

        # Compute weighted sum
        interaction_score = (0.6 * skill_match) + (0.2 * exp_score) + (0.2 * salary_score)
        interaction_matrix.loc[seeker_id, job_id] = interaction_score

# Convert matrix to float
interaction_matrix = interaction_matrix.astype(float)
interaction_matrix.head()
print("hello")
```

```python
from scipy.sparse.linalg import svds

# Assuming interaction_matrix is already computed
# Apply Singular Value Decomposition (SVD)
k = min(50, min(interaction_matrix.shape) - 1)  # Number of latent factors
U, sigma, Vt = svds(interaction_matrix, k=k)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)

# Reconstruct the predicted interaction matrix
predicted_matrix = np.dot(np.dot(U, sigma), Vt)

# Convert back to a DataFrame
predicted_df = pd.DataFrame(predicted_matrix, 
                            index=interaction_matrix.index, 
                            columns=interaction_matrix.columns)

# Show the top predicted job recommendations for each user
predicted_df.head()
```

```python

from scipy.sparse.linalg import svds
from scipy.sparse import csr_matrix  # Convert to sparse matrix

# Ensure all values are numeric (replace NaN with 0)
interaction_matrix = interaction_matrix.apply(pd.to_numeric, errors='coerce').fillna(0)

# Convert to sparse matrix format
interaction_sparse = csr_matrix(interaction_matrix.values)

# Apply Singular Value Decomposition (SVD)
k = min(50, min(interaction_matrix.shape) - 1)  # Number of latent factors
U, sigma, Vt = svds(interaction_sparse, k=k)

# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)

# Reconstruct the predicted interaction matrix
predicted_matrix = np.dot(np.dot(U, sigma), Vt)

# Convert back to a DataFrame
predicted_df = pd.DataFrame(predicted_matrix, 
                            index=interaction_matrix.index, 
                            columns=interaction_matrix.columns)

# Show the top predicted job recommendations for each user
predicted_df.head()
```

```python
np.savez("svd_model.npz", 
         U=U, 
         sigma=sigma, 
         Vt=Vt, 
         user_index=interaction_matrix.index.values, 
         job_index=interaction_matrix.columns.values)
```
