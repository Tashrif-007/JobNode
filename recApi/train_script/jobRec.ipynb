```python
import numpy as np
import pandas as pd
import re
import string
```

```python
df = pd.read_csv("../dataset/job_descriptions.csv")
df.head()
```

```python
df.shape
```

```python
columns_to_drop = [
    "Qualifications", "Country", "longitude", "latitude", "Work Type",
    "Company Size", "Contact", "Role", "Job Portal", "Benefits",
    "Responsibilities", "Company", "Company Profile","Preference","Contact Person","location","Job Posting Date"
]

df = df.drop(columns=columns_to_drop, errors='ignore')  # 'errors="ignore"' prevents errors if a column is missing

print(df.shape)  
df.head()
```

```python
# Extract min and max experience from the 'experience' column
df[['min_experience', 'max_experience']] = df['Experience'].str.extract(r'(\d+)\s*to\s*(\d+)')

# Convert to numeric (optional)
df['min_experience'] = df['min_experience'].astype(float)
df['max_experience'] = df['max_experience'].astype(float)

# Drop the original 'experience' column
df = df.drop(columns=['Experience'])

# Check the result # Verify the changed
df = df.drop(columns=['max_experience'])
df.head()
```

```python
# Extract min and max salary from the 'Salary Range' column
df[['min_salary', 'max_salary']] = df['Salary Range'].str.extract(r'(\d+)K?-?(\d+)K?')

# Convert to numeric (optional)
df['min_salary'] = df['min_salary'].astype(float) * 1000  # Convert 'K' to full value
df['max_salary'] = df['max_salary'].astype(float) * 1000  # Convert 'K' to full value

# Drop the original 'Salary Range' column
df = df.drop(columns=['Salary Range'])

# Check the result
df.head()
```

```python
df = df.rename(columns={'min_experience': 'experience'})
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['min_salary', 'max_salary']] = scaler.fit_transform(df[['min_salary', 'max_salary']])

# Check the result
df.head()
```

```python
# Function to clean text data
def clean_text(text):
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'\n', ' ', text)  # Remove newlines
    text = re.sub(r'\[.*?\]', '', text)  # Remove text in brackets
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words with numbers
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text.strip()

# Apply cleaning to text columns
text_columns = ['Job Description', 'skills','Job Title']
for col in text_columns:
    df[col] = df[col].apply(clean_text)

# Check the cleaned data
df.head()
```
